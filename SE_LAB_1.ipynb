multiple sets of inputs.



```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    coefficient_sets = [
        {'a': 0.4, 'b': -5, 'c': 40},
        {'a': 0.2, 'b': -2, 'c': 50},
        {'a': 0.15, 'b': -0.5, 'c': 25}
    ]

    time_values = np.linspace(0, 10, 100)

    for idx, coefficients in enumerate(coefficient_sets):
        temperature = quadratic_model(time_values, coefficients['a'], coefficients['b'], coefficients['c'])
        label = f'Set {idx + 1} coefficients'
        plt.plot(time_values, temperature, label=label)

        annotation_text = f'Set {idx + 1}: a={coefficients["a"]}, b={coefficients["b"]}, c={coefficients["c"]}'
        plt.annotate(annotation_text, xy=(time_values[-1], temperature[-1]), textcoords="offset points", xytext=(5,0), ha='left')

    plt.xlabel('time')
    plt.ylabel('temperature')
    plt.legend()
    plt.title('Weather Modeling with Quadratic Equation (Multiple Sets of Coefficients)')

    plt.show()

if __name__ == "__main__":
    main()
```

fixed and user input

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def plot_weather_model(time_values, temperature_values, label):
    plt.plot(time_values, temperature_values, label=label)

def main():
    # Example 1: Fixed coefficients
    time_values_fixed = np.linspace(0, 10, 100)
    a_fixed, b_fixed, c_fixed = 0.1, -1, 30
    temperature_fixed = quadratic_model(time_values_fixed, a_fixed, b_fixed, c_fixed)

    # Example 2: User-input coefficients
    a_user = float(input("Enter coefficient a: "))
    b_user = float(input("Enter coefficient b: "))
    c_user = float(input("Enter coefficient c: "))

    time_values_user = np.linspace(0, 10, 100)
    temperature_user = quadratic_model(time_values_user, a_user, b_user, c_user)

    # Plot the fixed coefficients
    plot_weather_model(time_values_fixed, temperature_fixed, 'Fixed Coefficients')

    # Plot the user-input coefficients on the same graph
    plot_weather_model(time_values_user, temperature_user, 'User Input Coefficients')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modeling with Quadratic Equation')
    plt.show()

if __name__ == "__main__":
    main()
```
